## Conditionals (If statements)

if x then:
    do a
if y then:
    do b
    
* > , <, ==,!= (not equal)

```python
mass = 3.78

if mass >= 3.0:
    print(mass, "is larger than 3")
if mass < 3.0:
    print(mass, "is smaller than 3")
```

```python
mass = 3.54
if mass != 3.0:
    print(mass, "is not 3")
if mass == 3.0:
    print(mass, "is 3")
```

### Use else to execute a block of code when an if condition is not true

```python
mass = 3.0
if mass > 3.0:
    print(mass, "is larger")
else:
    print(mass, "is smaller or equal")
```

```python
mass = 4.5

if mass > 3.0: #If the first condition is met, is will stop.
    print(mass, "is larger")
elif mass == 3.0: # You can only have one else.
    print(mass, "is 3")
else:
    print(mass, "is smaller")
```

```python
mass = 9.6

if mass > 9.0:
    print(mass, "is huge!")
elif mass > 3.0:
    print(mass, "is larger")
else:
    print(mass, "is smaller")
```

```python
grade = 85

if grade >= 90:
    print("Grade is A")
elif grade >= 80:
    print("Grade is B")
elif grade >= 70:
    print("Grade is C")
```

## For Loops

A for loop executes commands once for each value in a collection.

```python
temperature_01 = 17.5
temperature_02 = 15.9
temperature_03 = 19.4
temperature_04 = 21.1

print(temperature_01)
print(temperature_02)
print(temperature_03)
print(temperature_04)
```

```python
for temperature in [17.5, 15.9, 19.4, 21.1]: #temperature is the loop variable in the collection.
    print(temperature)
```

```python
temperatures = [17.5, 15.9, 19.4, 21.1]

for current_temperature in temperatures:
    print(current_temperature) #This is the body

print("hello")
print(current_temperature)
```

```python
primes = [2, 3, 5]

for number in primes:
    squared = number ** 2
    cubed = number ** 3
    print(number, squared, cubed)
```

### Use Range to iterate over a sequence of numbers

### Use range to repeat an action an arbitrary number of times

```python
for number in range(5):
    print("Again!")
```

## The Accumulator Pattern

```python
# Sum of the first 10 integers.

total = 0 # Accumulator variable 
for number in range(10):
    total = total + (number + 1)
    print(total)

print(total)
```

# Writing Functions


```python
def print_greeting():
    print("hello")

print_greeting()
```

```python
def print_date(year, month, day):
    joined = str(year) + "/" + str(month) + "/" + str(day)
    
    return joined

joined_date = print_date(1874, 3, 19)
print(joined_date)
```

```python
def average(values):
    if len(values) == 0:
        return None
    else:
        return sum(values) / len(values)
    
def average_2(values):
    if len(values) == 0:
        return None

    return sum(values) / len(values)

a = average_2([])
print(a)
    
```
